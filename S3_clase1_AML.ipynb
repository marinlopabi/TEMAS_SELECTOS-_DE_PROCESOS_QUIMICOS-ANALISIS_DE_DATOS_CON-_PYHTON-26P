# Errores comunes en Python

Antes de sumergirnos más, veamos algunos errores comunes que podemos encontrar en Python.

**Error de sintaxis (SyntaxError)**

Ocurre cuando el código no sigue las reglas de sintaxis de Python, como olvidar dos puntos después de una declaración de función o un bucle.

---
**Error de nombre (NameError)**

Ocurre cuando se hace referencia a una variable o función que no ha sido definida.

---
**Error de tipo (TypeError)**

Ocurre cuando se realiza una operación con tipos de datos incompatibles, como intentar sumar un número y una cadena.

---

**Error de índice (IndexError)**

Ocurre cuando se intenta acceder a un índice fuera del rango válido de una lista o secuencia.

## Hay dos formas de leer datos:
- local
- o desde URL

| **Función**        | **Descripción**                                                                                           |
|--------------------|---------------------------------------------------------------------------------------------------------------------|
| `read_csv()`       | Lee archivos delimitados por comas (CSV) desde disco, URL o buffer. Muy común para datos tabulares simples.        |
| `read_fwf()`       | Lee archivos de texto con columnas de ancho fijo. Útil cuando los datos no tienen delimitadores.                   |
| `read_clipboard()` | Lee datos copiados al portapapeles como si fueran un CSV. Ideal para copiar tablas desde Excel o páginas web.      |
| `read_excel()`     | Lee hojas de cálculo Excel (.xls, .xlsx). Soporta múltiples hojas. Usa `openpyxl` o `xlrd` como motores.            |
| `read_hdf()`       | Lee archivos HDF5 escritos con `pandas.HDFStore`. Eficiente para grandes volúmenes de datos jerárquicos.            |
| `read_html()`      | Extrae todas las tablas de una página HTML (como tablas de Wikipedia) y devuelve una lista de DataFrames.          |
| `read_json()`      | Lee archivos JSON o cadenas JSON válidas. Admite JSON anidados con `orient='records'` o `orient='split'`.           |
| `read_feather()`   | Lee archivos en formato Feather, optimizado para velocidad y compatibilidad con Apache Arrow.                       |
| `read_orc()`       | Lee archivos binarios en formato ORC (Optimized Row Columnar), usado en sistemas distribuidos como Hive.            |
| `read_parquet()`   | Lee archivos Parquet, un formato columnar eficiente muy usado en Big Data y Spark.                                 |
| `read_pickle()`    | Carga objetos serializados por `pandas.to_pickle()` con `pickle`. Útil para guardar DataFrames entre sesiones.     |
| `read_sas()`       | Lee archivos `.sas7bdat` o `.xpt` del software estadístico SAS.                                                     |
| `read_spss()`      | Carga archivos `.sav` de SPSS como DataFrames. Requiere `pyreadstat`.                                               |
| `read_sql()`       | Ejecuta consultas SQL sobre una conexión de base de datos (usando SQLAlchemy) y devuelve los resultados como DataFrame. |
| `read_sql_table()` | Carga una tabla completa desde una base de datos relacional vía SQLAlchemy, sin necesidad de escribir SQL manual.   |
| `read_stata()`     | Lee archivos `.dta` del software Stata. Puede importar etiquetas de variables y valores.                           |

## Archivos TXT

Un archivo txt se puede leer con read_csv() porque ambos son solo texto delimitado. Pandas no se fija en la extensión, sino en cómo están separados los datos. Si el txt tiene comas, punto y coma, tabulaciones o espacios, read_csv() sabe interpretarlo como tabla.

# Separadores en `pd.read_csv()`

En `pandas.read_csv()`, el parámetro `sep` indica **cómo están separados los datos** dentro de cada línea del archivo.

| Opción | ¿Qué significa? | ¿Cuándo usarla? | Ejemplo típico de línea |
|---|---|---|---|
| `sep=","` | Separador por **coma** (CSV clásico). | La mayoría de archivos `.csv`. | `A,B,C` |
| `sep=";"` | Separador por **punto y coma**. | Muy común en Excel configurado con coma decimal (LatAm/Europa). | `A;B;C` |
| `sep="\t"` | Separador por **tabulador** (TSV). | Archivos `.tsv` o exportaciones tipo “tabla”. | `A<TAB>B<TAB>C` |
| `sep="\s+"` | **Expresión regular**: uno o más espacios en blanco (espacios o tabs). | Archivos tipo “texto” donde los valores están separados por **varios espacios**. | `A   B     C` |
| `delim_whitespace=True` | Separar por **cualquier espacio en blanco** (espacios/tabs). | Similar a `sep="\s+"`, útil para tablas alineadas por espacios. | `A   B   C` |
| `sep=" "` | Separador por **un solo espacio**. | Solo si estás segura de que hay **exactamente un espacio** entre columnas. | `A B C` |
| `sep="|"` | Separador por **barra vertical** (pipe). | Logs o datos exportados como `|`. | `A|B|C` |
| `sep=":"` | Separador por **dos puntos**. | Archivos de configuración o formatos específicos. | `A:B:C` |
| `sep=","` + `decimal=","` | Coma como separador de columnas y **coma decimal**. | Datos europeos/latam donde decimales son `3,14`. | `3,14,2,71` (ojo con ambigüedad) |
| `sep=";"` + `decimal=","` | Punto y coma como separador y coma decimal. | MUY común en exportaciones de Excel en español. | `3,14;2,71;1,00` |
| `engine="python"` (cuando usas regex) | Permite separadores tipo regex como `\s+`. | Cuando `sep` es una **expresión regular**. | — |
| `comment="#"` | Ignora lo que va después de `#` en la línea. | Archivos con comentarios. | `A,B,C  # comentario` |
| `skipinitialspace=True` | Quita espacios **después** del separador. | Cuando queda `A, B, C` con espacios. | `A, B, C` |
| `thousands=","` | Interpreta separador de miles. | Números como `1,234,567`. | `1,234` |
| `quotechar='"'` | Define el caracter de comillas. | Si hay textos con comas dentro: `"CDMX, México"`. | `"CDMX, México",25` |

## Notas rápidas
- `sep="\s+"` y `delim_whitespace=True` sirven para separar por espacios/tabs.
- `sep="\s+"` es más flexible (regex), pero a veces requiere `engine="python"`.
- Si tus datos vienen de Excel en español, prueba primero: `sep=";"` y `decimal=","`.
