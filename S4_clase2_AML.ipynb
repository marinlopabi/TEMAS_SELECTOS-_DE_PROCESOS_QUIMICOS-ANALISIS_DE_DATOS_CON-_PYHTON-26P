# Análisis estadístico



El **análisis estadístico de datos** es el conjunto de técnicas que usamos para **describir, resumir e interpretar** datos, con el fin de **entender qué está pasando**, **comparar grupos** y, en algunos casos, **hacer inferencias o predicciones**.

---

## ¿Para qué sirve?
- **Resumir** grandes cantidades de datos en información clara.
- **Detectar patrones** (tendencias, diferencias entre grupos, relaciones).
- **Tomar decisiones** basadas en evidencia (por ejemplo, en ingeniería, economía, salud, educación).
- **Evaluar hipótesis**: comprobar si un cambio o una diferencia es “real” o puede explicarse por azar.

---

##  Tipos de análisis estadístico

### A) Estadística descriptiva
Responde preguntas como:
- ¿Cuál es el valor “típico”?
- ¿Qué tan dispersos están los datos?
- ¿Hay valores muy altos o muy bajos (outliers)?

Herramientas comunes:
- **Medidas de tendencia central:** media, mediana, moda
- **Medidas de dispersión:** rango, varianza, desviación estándar, IQR
- **Tablas y gráficos:** histogramas, boxplots, barras, pastel (según el caso)

### B) Estadística inferencial (saca conclusiones sobre una población)
Se usa cuando tienes una **muestra** y quieres concluir algo sobre una **población**.
Incluye:
- **Intervalos de confianza**
- **Pruebas de hipótesis** (p-values)
- **Comparaciones entre grupos** (t-test, ANOVA)
- **Relaciones entre variables** (correlación, regresión)

---

##  Pasos típicos de un análisis estadístico
1. **Plantear el objetivo**
   - Ejemplo: “Comparar si hay diferencia entre regiones”
2. **Revisar y limpiar datos**
   - valores nulos, errores, duplicados, formatos
3. **Explorar y describir**
   - tablas resumen y estadísticas básicas
4. **Visualizar**
   - gráficos para entender distribución y comparaciones
5. **Interpretar resultados**
   - qué significa en el contexto del problema
6. *(Opcional)* **Inferir o modelar**
   - pruebas estadísticas o modelos predictivos

---

##  Conceptos clave
- **Variable:** característica medida (ej. Total, Hombres, Mujeres)
- **Población vs muestra:** todos los datos posibles vs una parte
- **Distribución:** forma en la que se reparten los valores
- **Outliers:** valores atípicos que pueden afectar conclusiones

---

 **Idea central:**  
El análisis estadístico convierte datos en **información útil** para **describir** y **tomar decisiones** con fundamento.

## Paquetería `statistics` en Python

La paquetería **`statistics`** es un módulo **incluido en la librería estándar** de Python (no necesitas instalar nada) que sirve para realizar **estadística descriptiva básica** de manera rápida y clara.

---

### Permite:

### 1) Resumir datos experimentales

- **Media** (`mean`)
- **Mediana** (`median`)
- **Moda** (`mode`) y **multimoda** (`multimode`)

---

### 2) Medir variabilidad (dispersión)
Te ayuda a cuantificar qué tanto “se dispersan” tus mediciones:
- **Varianza muestral** (`variance`)
- **Desviación estándar muestral** (`stdev`)
- Versiones **poblacionales** (`pvariance`, `pstdev`)

Esto es muy útil para evaluar **precisión**, **ruido experimental** y **estabilidad del proceso**.

---

### 3) Calcular cuantiles y apoyar la detección de outliers
Con `quantiles()` puedes obtener cuartiles (Q1, Q2, Q3) para construir:
- **IQR (rango intercuartílico)**
- Criterio para **detectar valores atípicos (outliers)**




---

## Nota
- Cuando trabajas con **listas** pequeñas o medianas.
- Para análisis más avanzado (grandes datasets, DataFrames, gráficos, matrices), suele complementarse con **NumPy** y **Pandas**.


````python
import statistics as st
````

## Ejemplo: Nacimientos pág 10

````python
datos= [7.2, 7.8, 6.8, 6.2, 8.2, 8, 8.2, 5.6, 8.6, 7.1, 8.2, 7.7, 7.5, 7.2, 7.7, 5.8, 6.8, 6.8, 8.5, 7.5, 6.1,7.9, 9.4, 9, 7.8, 8.5, 9, 7.7, 6.7, 7.7 ]
````

# Medidas de tendencia central

- Media(promedio)
- Mediana
- Moda

````python
print("Media:", st.mean(datos))
print("Mediana:", st.median(datos))

# Moda: si no hay una moda clara, puede lanzar StatisticsError

print("Moda:", st.mode(datos))
````

## Medidas de dispersión
- Varianza
- DE
- Rango

````python
# Varianza y desviación estándar (muestrales)
print("Varianza (muestral):", st.variance(datos))
print("Std (muestral):", st.stdev(datos))

# poblacionales (si fuera TODA la población)
print("Varianza (poblacional):", st.pvariance(datos))
print("Std (poblacional):", st.pstdev(datos))


print("Rango:", max(datos) - min(datos))
````

## Cuartiles e IQR

````python
# n=4 devuelve tres cortes: Q1, Q2, Q3
q1, q2, q3 = st.quantiles(datos, n=4, method="inclusive")
iqr = q3 - q1

print("Q1, Q2, Q3:", q1, q2, q3)
print("IQR:", iqr)
````

## Detección de outliers con IQR

````python
li = q1 - 1.5 * iqr
ls = q3 + 1.5 * iqr
outliers = [x for x in datos if x < li or x > ls]

print("Límite inferior:", li)
print("Límite superior:", ls)
print("Outliers:", outliers)
````

## Coeficiente de variación (CV)



Indica **qué tan dispersos** están los datos **en relación con su promedio**.  
Se interpreta como un **porcentaje**: *“cuánta variación típica hay respecto a la media”*.


$$
CV = \frac{s}{\bar{x}} \times 100\% $$


$\bar{x} $= **media**

$s$ = **desviación estándar** (generalmente **muestral**)

---

### Interpretación
- **CV pequeño** → datos más **consistentes** (menor variación relativa).
- **CV grande** → datos más **dispersos** (mayor variación relativa).

> Ejemplo:  
> Si \(CV = 5\%\), la variación típica es aproximadamente el **5% del promedio**.


````python
cv = st.stdev(datos) / st.mean(datos) * 100
print("CV (%):", cv)
````

Nota: CV solo tiene sentido cuando la media es positiva y no está cerca de cero, porque si la media es muy pequeña el CV puede salir enorme y engañoso.

````python
import numpy as np
````

````python
a = np.array(datos)

n = a.size # tamaño del conjunto de datos
media = a.mean()
mediana = np.median(a) ###¿?

# Varianza y desviación estándar
var_pob = a.var(ddof=0)      # poblacional
std_pob = a.std(ddof=0)

var_mues = a.var(ddof=1)     # muestral
std_mues = a.std(ddof=1)

minimo = a.min()
maximo = a.max()
rango = maximo - minimo


print("media: ", media)
print("mediana: ", mediana)
print("varianza:", var_pob)
print("DE: ", std_pob)
print("rango: ", rango)
````

## Cuartiles y percentiles

````python
q1, q2, q3 = np.percentile(a, [25, 50, 75]) #  Cuartiles, percentiles e IQR
iqr = q3 - q1

li = q1 - 1.5 * iqr # Outliers por criterio IQR
ls = q3 + 1.5 * iqr
outliers = a[(a < li) | (a > ls)]


print("\nCuartiles:")
print(f"Q1 = {q1:.3f}, Q2 = {q2:.3f}, Q3 = {q3:.3f}")
print(f"IQR = {iqr:.3f}")
print("Outliers =", outliers)
````

## Coeficiente de variación (CV)

````python
cv = (std_mues / media) * 100
print(f"\nCoeficiente de variación (CV) = {cv:.2f}%")
````
