## Limpiar una lista

Con regex puedes:

- Buscar patrones en texto (encontrar).

- Validar si un texto cumple una forma (por ejemplo, un código, un email simple).

- Extraer partes de un texto (capturar).

- Reemplazar partes según un patrón (limpiar datos).

- Separar texto usando patrones (split “inteligente”).

| Función `re`                   | ¿Para qué sirve?                                                                | ¿Qué devuelve?                            | Ejemplo (patrón → texto)                           |
| ------------------------------ | ------------------------------------------------------------------------------- | ----------------------------------------- | -------------------------------------------------- |
| `re.search(patron, texto)`     | Busca la **primera coincidencia** en cualquier parte del texto                  | Un objeto *match* o `None`                | `r"\d+"` → `"Temp=35C"` (encuentra `35`)           |
| `re.match(patron, texto)`      | Busca coincidencia **solo al inicio** del texto                                 | Un objeto *match* o `None`                | `r"\d+"` → `"35C"` (sí), `"C35"` (no)              |
| `re.findall(patron, texto)`    | Encuentra **todas** las coincidencias                                           | Lista de strings (o tuplas si hay grupos) | `r"\d+"` → `"T1=10 T2=20"` → `["10","20"]`         |
| `re.finditer(patron, texto)`   | Itera sobre **todas** las coincidencias con más detalle (posición, grupo, etc.) | Iterador de *match objects*               | `r"\w+"` → `"Hola mundo"` (da matches con índices) |
| `re.sub(patron, repl, texto)`  | **Reemplaza** coincidencias (limpieza de texto, normalización)                  | String nuevo                              | `r"[#!?*]"` → `"#Hola!"` → `"Hola"`                |
| `re.subn(patron, repl, texto)` | Igual que `sub`, pero también dice **cuántos reemplazos** hizo                  | Tupla `(nuevo_texto, n)`                  | `r"\s+"` → `"a   b"` → `("a b", 1)`                |
| `re.split(patron, texto)`      | **Divide** un texto usando un patrón como separador                             | Lista de strings                          | `r"[,;]\s*"` → `"a, b; c"` → `["a","b","c"]`       |
| `re.compile(patron)`           | “Precompila” un patrón para reutilizarlo (más limpio y a veces más rápido)      | Objeto `Pattern`                          | `p = re.compile(r"\d+")`; luego `p.findall(texto)` |
| `re.escape(texto)`             | Escapa caracteres especiales para tratarlos como literales                      | String escapado                           | `"a+b"` → `"a\+b"` (para buscar `a+b` literal)     |
| `re.fullmatch(patron, texto)`  | Verifica si **todo el texto** coincide con el patrón                            | Un objeto *match* o `None`                | `r"\d+"` → `"123"` (sí), `"123a"` (no)             |



Ejemplo:

 A partir de la lista capitales, crea una nueva lista llamada **capitales_limpias** donde cada elemento quede sin los símbolos # * ! ?, sin espacios al inicio o al final, y con un formato consistente (por ejemplo, cada palabra iniciando con mayúscula).

 No modifiques la lista original; genera una nueva.


    ¿Qué hacer?
    1) Convierte a string (evita errores si hay None o números).
    2) Quita espacios al inicio y al final.
    3) Elimina símbolos no deseados: ! * # ?
    4) Normaliza espacios internos (muchos espacios -> uno).
    5) Capitaliza palabra por palabra con control:
       - Conectores comunes en español van en minúsculas: de, del, la, el, y...
       - Si el conector es la primera palabra, se deja capitalizado.

```python
import re
def limpiar_cadenas(strings):
    result = []
    for value in strings:
        value = value.strip()  # suprime blancos u otro carácter especificado del final
                               #o del principio de una expresión
        value = re.sub('[!*#?]', '', value) #reemplaza todas las ocurrencias de un patrón
                                           #por el texto que especifiquemos
        value = value.title() #convierte la primera letra de cada palabra de una cadena
                              # a mayúsculas.
        result.append(value)
    return result
```

```python
capitales_limpias = limpiar_cadenas(capitales)

capitales_limpias.sort()
```
