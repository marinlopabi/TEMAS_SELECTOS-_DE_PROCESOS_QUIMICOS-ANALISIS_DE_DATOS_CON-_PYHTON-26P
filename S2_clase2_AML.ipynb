# Tuplas

Una tupla es una estructura de datos inmutable y ordenada que permite almacenar una colección de elementos. Los elementos de una tupla se encierran entre paréntesis (), separados por comas.

- Para crear una tupla, encierra los elementos entre paréntesis:

```python
vector = (3, 4, 1)
```

Para acceder a los elementos de una tupla, utiliza el índice del elemento entre corchetes, similar a las listas:

```python
print(vector[1])  # Imprime la posición 1 de la tupla
```

A diferencia de las listas, las tuplas son inmutables, lo que significa que no se pueden modificar una vez creadas. No se pueden agregar, eliminar o cambiar elementos en una tupla existente.

Las tuplas son útiles cuando necesitas almacenar una colección de elementos que no deben modificarse, como coordenadas o datos de configuración.

## Métodos de tuplas

Aunque las tuplas son inmutables, Python proporciona varios métodos útiles para trabajar con ellas:

**count(elemento)**: devuelve el número de veces que aparece un elemento en la tupla.

**index(elemento)**: devuelve el índice de la primera aparición de un elemento en la tupla. Opcionalmente, se puede especificar el inicio y fin de la búsqueda.

**len(tupla)**: aunque no es un método de tupla propiamente dicho, esta función incorporada devuelve la longitud de la tupla.

```python
mi_tupla = (1, 2, 3, 2, 4, 2)


print (mi_tupla.index(2))

print (mi_tupla.index(2, 2))

print (mi_tupla.index(2, 2, 4))
```

# Diccionarios

Un **diccionario** es una estructura de datos que permite **guardar información en pares**:
**clave → valor**.

- A diferencia de las listas (que usan posiciones), los diccionarios usan **nombres (claves)** para acceder a los datos.




Un diccionario funciona como:
- un **diccionario real** (palabra → significado)
- o una **tabla con etiquetas**


## Métodos de diccionarios

Los diccionarios en Python tienen varios métodos incorporados para manipular y acceder a los elementos. Algunos métodos comunes son:

- **keys()**: devuelve una vista de todas las claves del diccionario.
- **values()**: devuelve una vista de todos los valores del diccionario.
- **items()**: devuelve una vista de todos los pares clave-valor del diccionario.
- **update(otro_diccionario)**: actualiza el diccionario con los pares clave-valor de otro diccionario.

Una forma de crear un diccionario es utilizando llaves "{}" y signos de dos puntos ":" para separar claves y valores:

```python
reactor = {
    "tipo": "CSTR",
    "volumen": 5.0,
    "temperatura": 350,
    "presion": 2.5
}
reactor
```

Se puede acceder a los elementos, insertarlos o formar conjuntos con ellos utilizando la misma sintaxis que para acceder a los elementos de una lista:

```python
print(reactor["temperatura"])
```

```python
reactor["temperatura"] = 360 #Modifica la temperatura
reactor["conversion"] = 0.75 #Agrega un nuevo elemento
```

```python
print(reactor)
```

¿Para qué usar diccionarios?

En Ingeniería Química son ideales para:

- Guardar propiedades de un sistema

- Almacenar parámetros de un proceso

- Organizar resultados por nombre

- Evitar confusiones con posiciones

```python
propiedades = {
    "densidad": 998,
    "viscosidad": 0.001,
    "Cp": 4.18
}
print(propiedades)
```

| Lista                      | Diccionario                 |
| -------------------------- | --------------------------- |
| Usa índices (0,1,2)        | Usa claves                  |
| Ordenada                   | No depende del orden        |
| Ideal para datos repetidos | Ideal para datos con nombre |


# Funciones en Python

Las **funciones** son el método **más importante** en Python para **organizar**, **reutilizar** y **simplificar** el código.

Regla general:  
Si un mismo código se va a usar **más de una vez**, conviene escribir una **función**.

---

## 1. ¿Qué es una función?

Una **función** es un bloque de código que:
- Realiza una tarea específica
- Tiene un nombre
- Puede recibir datos de entrada
- Puede devolver un resultado


Las funciones permiten:
- Evitar repetir código
- Hacer el programa más claro
- Facilitar correcciones y cambios
- Organizar cálculos complejos



En Python, las funciones se declaran con la palabra clave **`def`**, contiene un bloque de código, con un uso opcional de la palabra clave return.



```python
def sumar(a, b): #Lo que se encuentra dentro de los parentesis son los parámetros de entrada
    x = a + b
    return x
```

¿Cómo usar la función?

```python
sumar(3,5)
```

Las funciones pueden recibir información.

```python
def imprimir_temperatura(T):
    print("La temperatura es:", T)
```

```python
imprimir_temperatura(350)
```

Ejemplo: Función para obtener la conversión

```python
def calcular_conversion(A0, A):
    conversion = (A0 - A) / A0
    return conversion
```

```python
X = calcular_conversion(10, 3)
print(X)
```

### Funciones con un número variable de argumentos (*).

```python
def suma_variable(*numeros):
    total = 0
    for numero in numeros:
        total += numero
    return total



print(suma_variable(4, 5, 6, 7))
```

Código para calcular el promedio de datos

```python
def promedio(*datos):
  suma = sum(datos)
  total = len(datos)
  res = suma / total
  return res
```

```python
promedio(3,5,6,7)
```

## Funciones anónimas (lambda)

Python permite crear funciones anónimas o funciones lambda, que son funciones sin nombre definidas en una sola línea. Se utilizan comúnmente para funciones pequeñas y concisas

```python
def sumar_2(x):
    sumar= x+2
    return sumar
```

```python
sumar_2(3)
```

```python
sumar =lambda x: x + 2
print(sumar(3))
```

## Documentación de funciones (docstrings)

Es una buena práctica documentar nuestras funciones utilizando docstrings.

Los docstrings son cadenas de texto que describen el propósito, los parámetros y el valor de retorno de una función. Se colocan inmediatamente después de la definición de la función y se encierran entre triples comillas dobles

```python
def area_rectangulo(base, altura):
    """
    Calcula el área de un rectángulo.


    Args:
        base (float): La base del rectángulo.
        altura (float): La altura del rectángulo.


    Returns:
        float: El área del rectángulo.
    """
    return base * altura
```

Buenas prácticas:

- Usar nombres claros
- Una función = una tarea
- Evitar funciones muy largas
- Usar return cuando se necesita un resultado
